```python
# df=replicas_copy.DataFrame()
import re
pt=r'еня\sзовут\s\w+\s'
pt2=r'меня\s\w+\sзовут'
pt3=r'компания\s\w+'
list_greeting=[]
list_introduce=[]
list_name_manager=[]
list_name_company=[]
list_goodbye=[]

for index, row in replicas_copy.iterrows():
    if row['role']=='manager' and ' здравствуйте' in row['text']  :   
        '''Pеплики с приветствием – где менеджер поздоровался'''
        list_greeting.append(row['text'])
#         print(row['role'],row['text'])

        
    if row['role']=='manager' and 'Меня зовут' in row['text']  :   
        '''Pеплики, где менеджер представил себя'''
        list_introduce.append(row['text'])
     
        
        
    if row['role']=='manager' and  re.findall(pt,row['text'])    :   
#          '''Имя менеджера'''
        a=re.findall(pt,row['text'])
        a=a[0][10:]
        list_name_manager.append(a)
#         print(a[0][10:])  

            
    if  re.findall(pt2,row['text']):
#          '''Имя менеджера'''
        b=re.findall(pt2,row['text'])
        b=b[0][5:11]
        list_name_manager.append(b)
#         print(b[0][5:11]) 
        
        
        
    if row['role']=='manager' and 'компани' in row['text']  :   
        if re.findall(pt3,row['text']):
            c=re.findall(pt3,row['text'])
            
            c=c[0][9:19]
            list_name_company.append(c)
#             print(c)

    if row['role']=='manager' and ' свидания' in row['text']  :   
        '''Pеплики где менеджер попрощался.'''
        list_goodbye.append(row['text'])    

                

list_name_company.pop(-1)

print('a. Извлекать реплики с приветствием – где менеджер поздоровался.  \n',list_greeting )

print('b. Извлекать реплики, где менеджер представил себя. \n',list_introduce)
print('c. Извлекать имя менеджера.\n',list_name_manager)
print('d. Извлекать название компании.\n',list_name_company)
print('e.Извлекать реплики, где менеджер попрощался.',list_goodbye)

```
